## XML example

+ for details about tree traversal and iterators, see https://docs.python.org/2.7/library/xml.etree.elementtree.html

```python
document_tree = ET.parse( './data/mondial_database_less.xml' )
```

```python
# print names of all countries
for child in document_tree.getroot():
    print(child.find('name').text)
```

```python
# print names of all countries and their cities
for element in document_tree.iterfind('country'):
    print('* ' + element.find('name').text + ':'),
    capitals_string = ''
    for subelement in element.getiterator('city'):
        capitals_string += subelement.find('name').text + ', '
    print(capitals_string[:-2])
```

Using data in 'data/mondial_database.xml', the examples above, and refering to https://docs.python.org/2.7/library/xml.etree.elementtree.html, find

```python
data = []
for child in document.getroot():
    for child2 in child:
        if child2.tag == "infant_mortality":
            y = child2.text
        x = child.find('name').text
    if child.tag == 'country':
        row = (x, y)
        data.append(row)
data
```

#### 1) 10 countries with the lowest infant mortality rates

```python
inf_mortality_df = pd.DataFrame.from_records(data, columns = ['name', 'infant_mortality'])
inf_mortality_df['infant_mortality'] = inf_mortality_df['infant_mortality'].astype('float64')
inf_mortality_df.sort_values('infant_mortality').head(10)
```

#### 2) 10 cities with the largest population

```python
data = []
for child in document.getroot():
    for child2 in child:
        if child2.tag == 'city':
            for child3 in child2:
                year = 0
                if child3.tag == 'name':
                    x = child3.text
                if child3.tag == 'population':
                    if int(child3.attrib['year']) > year:
                        y = child3.text
            row = (x, y)
            data.append(row)
data
                        
                    #print(child3.tag, child3.attrib, child3.text)
```

```python
cities_df = pd.DataFrame.from_records(data, columns = ['name', 'population'])
cities_df['population'] = cities_df['population'].astype('float64')
cities_df.sort_values('population', ascending = False).head(10)
```

#### 3) 10 ethnic groups with the largest overall populations (sum of best/latest estimates over all countries)

```python
ethnic_groups = []
for child in document.getroot():
    for child2 in child:
        if child2.tag == 'population':
            country_pop = int(child2.text)
        if child2.tag == 'ethnicgroup':
            group = child2.text
            dic = child2.attrib
            group_prcnt = float(dic['percentage'])/100
            group_pop = round(country_pop * group_prcnt) 
            ethnic_groups.append((group, group_pop))
ethnic_groups
    
```

```python
ethnic_group_df = pd.DataFrame.from_records(ethnic_groups, columns = ['groupname', 'population'])
ethnic_group_df.sort_values('population', ascending = False).head(10)
```

#### 4) name and country of a) longest river, b) largest lake and c) airport at highest elevation

```python
for child in document.getroot():
    if child.tag == ('river' or 'lake' or 'airport'):
        print(child.tag, " = parent")
        for child2 in child:
            print(('<{}> {}').format(child2.tag, child2.text))
```

```python
landmark_data = []
for child in document.getroot():
    if child.tag == 'river' or child.tag == 'lake' or child.tag == 'airport':
        _type = child.tag
        for child2 in child:
            if child2.tag == 'name':
                name = child2.text
            if child2.tag in ['length', 'a rea', 'elevation']:
                trait = child2.tag
                try:
                    value = float(child2.text)
                except TypeError:
                    value = None
                landmark_data.append((_type, name, trait, value))
landmark_data
```

```python
landmark_df = pd.DataFrame.from_records(landmark_data, columns = ['_type', 'name', 'trait', 'value'])
landmark_df.head(10)
```

```python
top_df = pd.DataFrame(columns = ['_type', 'name', 'trait', 'value'])
top_df = top_df.append((landmark_df[landmark_df['_type'] == 'river']).sort_values(by = 'value', ascending = False).head(1))
top_df = top_df.append((landmark_df[landmark_df['_type'] == 'airport']).sort_values(by = 'value', ascending = False).head(1))
top_df = top_df.append((landmark_df[landmark_df['_type'] == 'lake']).sort_values(by = 'value', ascending = False).head(1))
top_df
```
